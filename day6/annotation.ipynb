```python
# 어노테이션 : 어떤 타입인지 미리 적어두는 표시

age = int = 30
print(age)

# 어노테이션은 강제 규칙이 아니다.
age : int = "안녕"
print(age)

# 대규모 프로젝트에서는 미리 적어둬야 버그를 줄일 수 있음
def add(a: int, b: int) -> int:
    return a + b

add(10, 20)

add('안녕', '반가워')

numbers: list[int] = [1, 2, 3]

person : dict[str, int] = {
    'age' : 30,
    'height' : 170
}

print(numbers)
print(person)

```
