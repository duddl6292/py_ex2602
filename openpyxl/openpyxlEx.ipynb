```python
from openpyxl import Workbook, load_workbook

# 새로운 엑셀 파일 만들기
# 새 엑셀 파일을 만들겠다 라는 의미
w1 = Workbook()

# 만들어진 엑셀 파일에서 기본 시트를 선택
ws = w1.active

# 지금 만든 엑셀 파일을 실제 파일로 저장
# 파일의 이름은 test.xlsx
w1.save('test.xlsx')

w2 = load_workbook('test.xlsx')

ws2 = w2.active

# 셀에 값 입력하기
# A열 1행 셀에 hello라는 텍스트 입력
ws2['A1'] = 'hello'

ws2['c3'] = 'by'

# 변경사항 저장
w2.save('test.xlsx')

a = 'python'

num = 2
# A1부터 C3까지 범위 선택(3행 * 3열 = 9개의 셀)
for row in ws2['A1':'C3']:
    for cell in row:
        cell.value = num
        num += 2
 
```

```python
import openpyxl

item1 = {
    '품목' : '바나나',
    '가격' : '30000',
    '수량' : 1,
    '날짜' : '2026년 2월 20일'   
}

item2 = {
    '품목' : '컴퓨터',
    '가격' : '3000000',
    '수량' : 3,
    '날짜' : '2026년 2월 21일'   
}

item3 = {
    '품목' : '커피',
    '가격' : '3000',
    '수량' : 5,
    '날짜' : '2026년 2월 22일'   
}

item4 = {
    '품목' : '마우스',
    '가격' : '150000',
    '수량' : 3,
    '날짜' : '2026년 2월 23일'   
}

wb = Workbook()
ws = wb.active


ws['A1'] = '품목' # A열 1행 : 품목
ws['B1'] = '가격'
ws['C1'] = '수량'
ws['D1'] = '날짜'

items = [item1, item2, item3, item4]

i = 2
for item in items:
    ws['A'+str(i)] = item['품목']
    ws['B'+str(i)] = item['가격']
    ws['C'+str(i)] = item['수량']
    ws['D'+str(i)] = item['날짜']
    i += 1
    
wb.save('test2.xlsx')
```
