```python
def get_1_to_n_prime_sum(n):
    sum = 0
    for i in range(2, n+1):
        count = 0
        for j in (1, i+1):
            if i % j == 0:
                count += 1
        if count == 2:
            sum += i
    return(sum)

print(get_1_to_n_prime_sum(100))
```

```python
def get_divisors_sum(num):
    if num == 1:
        return False
    i = 2
    while i < num:
        if num % i == 0:
            return False
        i += 1
    return True
get_divisors_sum(3)
```

```python
# dictionary
ages = {
    "철수": 10,
    "영희": 20,
    "영수": 30,
}
ages
print("== 딕셔너리 반복 v1 ==")
for name in ages:
    print("{} 나이 : {}".format(name, ages[name]))

print("== 딕셔너리 반복 v2 ==")
print(ages.keys())
for name in ages.keys():
    print("{} 나이 : {}".format(name, ages[name]))
print("== 딕셔너리 반복 v3 ==")
for name in ages.values():
    print("나이 : {}".format(name))
print("== 딕셔너리 반복 v4 ==")
for name, age in ages.items():
    print("{} 나이 : {}".format(name, age))    
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담기
month = {
    '1월': 31,
    '2월': 28,
    '3월': 31,
    '4월': 30,
    '5월': 31,
    '6월': 30,
    '7월': 31,
    '8월': 31,
    '9월': 30,
    '10월': 31,
    '11월': 30,
    '12월': 31,
}
print(month)

for m in month:
    print("{} {}일".format(m, month[m]))
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 반복문을 통해 담기
month = {}
for m in range(1, 13):
    if m == 2:
        month[m] = 28
    elif m in [4, 6, 9, 11]:
        month[m] = 30
    else :
        month[m] = 31
for i in month:
    print("{}월 {}일".format(i, month[i]))
```
