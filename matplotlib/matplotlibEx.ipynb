```python
fig = plt.figure()

# (행개수, 열개수, 위치값)
a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(1, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, 'rd')

x = range(0, 50)
y = range(0, 50)
a2.plot(x, y, 'y.')

# x축, y축 이름 지정
a1.set_xlabel('time')
a1.set_ylabel('money')

a2.set_xlabel('time')
a2.set_ylabel('money')


plt.show()
```

```python
fig  = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.bar(x, y, tick_label=labels, color=colors) # bar뒤에 h를 붙이면 가로로 나옴(barh)

plt.show()

```

```python
fig  = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.bar(x, y, tick_label=labels, color=colors) 

for i, v in enumerate(y):
    a.text(v + 2, i + 1.05, str(v)) # 표준 수치 구하는 공식

plt.show()

```
