```python
#파이썬 기본 자료형(int, float, str, bool(T or F))
age = 26
height = 182.5
name = "김호영"
is_exercise = True

print(type(age))
print(type(height))
print(type(name))
print(type(is_exercise))
```

```python
#파이썬 컬렉션 자료형 (list, dict)
weight = [60, 65, 70, 75]
print(weight)
print(type(weight))

person = {
    "weight" : 65,
    "name" : "김호영",
    }
print(person)
print(type(person))

person_info = (40, 172, 68)
blood_pressures = {120, 162}
```

```python
#자료형 변환
#문제1
a = 10
b = 20

a = str(a)
b = str(b)

print("문제1 : {}".format(a+b))

#문제2
a = '3'
b = '6'

a = int(a)
b = int(b)

print("문제2 : {}".format(a + b))

#문제3
a = '3.1'
b = '6.1'

a = float(a)
b = float(b)

print("문제3 : {}".format(a + b))

#문제4
a = '3.1'
b = '6.1'

a = int(float(a))
b = int(float(b))

print("문제4 : {}".format(a + b))
```
