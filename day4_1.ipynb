```python
# 클래스 -> 객체를 표현하기 위한 문법의 일종, 
# 프로그래밍으로 객체를 만들 때 사용하는 것

# 전사, 마법사, 궁수, 자동차, 비행기 -> object(객체)
# -> 현실, 가상, 개념

# 속성(명사적) - attribute
# -> 
# 능력(기능, 동사적) - mrthod

class Person:
    def hello(self):
        print("hi")

hy = Person() # hy는 Person의 인스턴스
hy.hello()

b = list(range(10))
print(b)

a = int(10)
print(type(a))

b = [1, 2, 3]
print(type(b))

c = {'a':1, 'b':2}
print(type(c))
```
