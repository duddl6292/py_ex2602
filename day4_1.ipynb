```python
class Person():
    def __init__(self):              
        self.hello = '안녕하세요'     # instance의 attribute # 속성이란 객체 안에 저장된 변수 

    def hi(self):                    # method
        print(self.hello)

jin = Person()      # 이 때 __init__ 메서드는 이미 실행됐다
jin.hi()
```

```python
class Person():
    def __init__(self, name, age, hobby):              
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby     
# __init__안에 저장해놓는걸 인스턴스 속성이라고 함
    def intoduce(self):                
        print("{}, 저는 {}입니다.".format(self.hello, self.name))

jin = Person('진', 22, '운동')      
jin.intoduce()

print('취미 : {}'.format(jin.hobby))
```

```python
# 클래스 외부에서는 접근 불가, 클래스 내부에서만 사용 가능한 private attribute
class Person():
    def __init__(self, name, age, hobby, weight):              
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby 
        self.__weight = weight

jin = Person('진', 22, '운동', 78) 
jin.__weight -= 20

print(jin.__weight)
```

```python
# 클래스 외부에서는 접근 불가, 클래스 내부에서만 사용 가능한 private attribute
class Person():
    def __init__(self, name, age, hobby, weight):              
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby 
        self.__weight = weight

    def diet(self, loss_weight):
        self.__weight -= loss_weight
        print("몸무게 : ", self.__weight)

jin = Person('진', 22, '운동', 78) 
jin.diet(20)
```
